```python
import torch
from torch import nn
import torch.nn.functional as F

class IDNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc_1 = nn.Linear(256,128)
        self.batchnorm_1 = nn.BatchNorm1d(128)
        self.l_relu_1 = nn.LeakyReLU(negative_slope=0.2)
        self.fc_2 = nn.Linear(128,64)
        self.batchnorm_2 = nn.BatchNorm1d(64)
        self.l_relu_2 = nn.LeakyReLU(negative_slope=0.2)
        self.fc_3 = nn.Linear(64,32)
        self.batchnorm_3 = nn.BatchNorm1d(32)
        self.l_relu_3 = nn.LeakyReLU(negative_slope=0.2)
        # self.fc_5 = nn.Linear(32,16)
        # self.batchnorm_5 = nn.BatchNorm1d(16)
        # self.l_relu_5 = nn.LeakyReLU(negative_slope=0.2)
        # self.fc_6 = nn.Linear(16,32)
        # self.batchnorm_6 = nn.BatchNorm1d(32)
        # self.l_relu_6 = nn.LeakyReLU(negative_slope=0.2)
        self.fc_7 = nn.Linear(32,64)
        self.batchnorm_7 = nn.BatchNorm1d(64)
        self.l_relu_7 = nn.LeakyReLU(negative_slope=0.2)



    def forward(self,x):
        x = self.fc_1(x)
        x = self.batchnorm_1(x)
        x = self.l_relu_1(x)
        x = self.fc_2(x)
        x = self.batchnorm_2(x)
        x = self.l_relu_2(x)
        x = self.fc_3(x)
        x = self.batchnorm_3(x)
        x = self.l_relu_3(x)
        # x = self.fc_5(x)
        # x = self.batchnorm_5(x)
        # x = self.l_relu_5(x)
        # x = self.fc_6(x)
        # x = self.batchnorm_6(x)
        # x = self.l_relu_6(x)
        x = self.fc_7(x)
        # x = self.batchnorm_7(x)
        # x = self.l_relu_7(x)
        return x


class Conv1d(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size, stride, padding):
        super(Conv1d, self).__init__()
        self.conv1 = nn.Conv1d(in_channels, out_channels,
                              kernel_size,stride, padding)
        self.batchnorm1 = nn.BatchNorm1d(out_channels)
        self.l_relu1 = nn.LeakyReLU(negative_slope=0.2)
        self.conv2 = nn.Conv1d(out_channels, out_channels,
                              kernel_size, stride,padding)
        self.batchnorm2 = nn.BatchNorm1d(out_channels)
        self.l_relu2 = nn.LeakyReLU(negative_slope=0.2)

    def forward(self, x):
        x = self.conv1(x)
        x = self.batchnorm1(x)
        x = self.l_relu1(x)
        x = self.conv2(x)
        x = self.batchnorm2(x)
        x = self.l_relu2(x)

        return x

class FC(nn.Module):
    def __init__(self, in_dim, out_dim):
        super(FC, self).__init__()
        self.fc_1 = nn.Linear(in_dim,out_dim)
        self.batchnorm_1 = nn.BatchNorm1d(out_dim)
        self.l_relu_1 = nn.LeakyReLU(negative_slope=0.2)

    def forward(self, x):
        x = self.fc_1(x)
        x = self.batchnorm_1(x)
        x = self.l_relu_1(x)

        return x


class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        filter_ = 16
        self.filter_ = filter_
        self.half = int(filter_/2)
        self.conv1 = Conv1d(in_channels = 1,out_channels = filter_,kernel_size = 5,stride = 1,padding = 2)
        self.pool1 = nn.MaxPool1d(kernel_size = 2,stride = 2)
        self.conv2 = Conv1d(in_channels = filter_,out_channels = filter_ * 2,kernel_size = 5,stride = 1,padding = 2)
        self.pool2 = nn.MaxPool1d(kernel_size = 2,stride = 2)
        self.conv3 = Conv1d(in_channels = filter_ * 2,out_channels = filter_ * 4,kernel_size = 5,stride = 1,padding = 2)
        self.pool3 = nn.MaxPool1d(kernel_size = 2,stride = 2)
        self.conv4 = Conv1d(in_channels = filter_ * 4,out_channels = filter_ * 8,kernel_size = 5,stride = 1,padding = 2)
        self.pool4 = nn.MaxPool1d(kernel_size = 2,stride = 2)
        self.conv5 = Conv1d(in_channels = filter_ * 8,out_channels = filter_ * 16,kernel_size = 5,stride = 1,padding = 2)
        # self.fc_1 = FC(filter_ * 8*dim,256)
        # self.fc_2 = FC(256,filter_ * 8*dim)
        #  self.t_conv1 = nn.ConvTranspose2d(in_channels = filter_ * 16,out_channels = filter_ * 8,kernel_size = 2,stride = 2)
        # self.conv6 = Conv1d(in_channels = filter_ * 8,out_channels = filter_ * 8,kernel_size = 3,stride = 1,padding = 1)
        self.t_conv1 = nn.ConvTranspose1d(in_channels = filter_ * 16,out_channels = filter_ * 16,kernel_size = 2,stride = 2)
        self.conv6 = Conv1d(in_channels = filter_ * 16,out_channels = filter_ * 8,kernel_size = 5,stride = 1,padding = 2)
        self.t_conv2 = nn.ConvTranspose1d(in_channels = filter_ * 8,out_channels = filter_ * 8,kernel_size = 2,stride = 2)
        self.conv7 = Conv1d(in_channels = filter_ * 8,out_channels = filter_ * 4,kernel_size = 5,stride = 1,padding = 2)
        self.t_conv3 = nn.ConvTranspose1d(in_channels = filter_ * 4,out_channels = filter_ * 4,kernel_size = 2,stride = 2)
        self.conv8 = Conv1d(in_channels = filter_ * 4,out_channels = filter_ * 2,kernel_size = 5,stride = 1,padding = 2)
        self.conv9 = Conv1d(in_channels = filter_ * 2,out_channels = filter_ ,kernel_size = 5,stride = 1,padding = 2)
        self.conv10 = nn.Conv1d(in_channels = filter_,out_channels = 1,kernel_size = 5,stride = 1,padding = 2)


    def forward(self,x):
        batch = x.shape[0]
        x = x.view(batch,1,-1)
        x = self.conv1(x)
        x = self.pool1(x)
        x = self.conv2(x)
        x = self.pool2(x)
        x = self.conv3(x)
        x = self.pool3(x)
        x = self.conv4(x)
        x = self.pool4(x)
        x = self.conv5(x)
        x = self.t_conv1(x)
        x = self.conv6(x)
        x = self.t_conv2(x)
        # print(x.shape)
        x = self.conv7(x)
        x = self.t_conv3(x)
        x = self.conv8(x)
        x = self.conv9(x)
        # print(x.shape)
        x = self.conv10(x)
        x = x.view(batch,-1)

        return x


class CAE(nn.Module):
    def __init__(self):
        super().__init__()
        filter_ = 16
        self.filter_ = filter_
        self.half = int(filter_/2)
        self.conv1 = Conv1d(in_channels = 1,out_channels = filter_,kernel_size = 5,stride = 1,padding = 2)
        self.pool1 = nn.MaxPool1d(kernel_size = 2,stride = 2)
        self.conv2 = Conv1d(in_channels = filter_,out_channels = filter_ * 2,kernel_size = 5,stride = 1,padding = 2)
        self.pool2 = nn.MaxPool1d(kernel_size = 2,stride = 2)
        self.conv3 = Conv1d(in_channels = filter_ * 2,out_channels = filter_ * 4,kernel_size = 5,stride = 1,padding = 2)
        self.pool3 = nn.MaxPool1d(kernel_size = 2,stride = 2)
        self.conv4 = Conv1d(in_channels = filter_ * 4,out_channels = filter_ * 8,kernel_size = 5,stride = 1,padding = 2)
        self.pool4 = nn.MaxPool1d(kernel_size = 2,stride = 2)
        self.conv5 = Conv1d(in_channels = filter_ * 8,out_channels = filter_ * 16,kernel_size = 5,stride = 1,padding = 2)
        self.fc_1 = FC(filter_ * 128,16)
        self.fc_2 = FC(16,filter_ * 128)
        #  self.t_conv1 = nn.ConvTranspose2d(in_channels = filter_ * 16,out_channels = filter_ * 8,kernel_size = 2,stride = 2)
        self.conv6 = Conv1d(in_channels = filter_ * 16,out_channels = filter_ * 8,kernel_size = 3,stride = 1,padding = 1)
        self.t_conv1 = nn.ConvTranspose1d(in_channels = filter_ * 8,out_channels = filter_ * 8,kernel_size = 2,stride = 2)
        self.conv7 = Conv1d(in_channels = filter_ * 8,out_channels = filter_ * 4,kernel_size = 5,stride = 1,padding = 2)
        self.t_conv2 = nn.ConvTranspose1d(in_channels = filter_ * 4,out_channels = filter_ * 4,kernel_size = 2,stride = 2)
        self.conv8 = Conv1d(in_channels = filter_ * 4,out_channels = filter_ * 2,kernel_size = 5,stride = 1,padding = 2)
        self.t_conv3 = nn.ConvTranspose1d(in_channels = filter_ * 2,out_channels = filter_ * 2,kernel_size = 2,stride = 2)
        self.conv9 = Conv1d(in_channels = filter_ * 2,out_channels = filter_ ,kernel_size = 5,stride = 1,padding = 2)
        self.t_conv4 = nn.ConvTranspose1d(in_channels = filter_ ,out_channels = filter_ ,kernel_size = 2,stride = 2)
        self.conv10 = Conv1d(in_channels = filter_ ,out_channels = filter_ ,kernel_size = 5,stride = 1,padding = 2)
        self.conv11 = nn.Conv1d(in_channels = filter_,out_channels = 1,kernel_size = 5,stride = 1,padding = 2)


    def forward(self,x):
        batch = x.shape[0]
        x = x.view(batch,1,-1)
        x = self.conv1(x)
        x = self.pool1(x)
        x = self.conv2(x)
        x = self.pool2(x)
        x = self.conv3(x)
        x = self.pool3(x)
        x = self.conv4(x)
        x = self.pool4(x)
        # print(x.shape)
        x = self.conv5(x)
        # print(x.shape)
        x = x.view(batch,-1)
        # print(x.shape)
        x = self.fc_1(x)
        # print(x.shape)
        x = self.fc_2(x)
        x = x.view(batch,self.filter_*16,-1)

        x = self.conv6(x)
        x = self.t_conv1(x)
        x = self.conv7(x)
        x = self.t_conv2(x)
        # print(x.shape)
        x = self.conv8(x)
        x = self.t_conv3(x)
        x = self.conv9(x)
        x = self.t_conv4(x)
        # print(x.shape)
        x = self.conv10(x)
        x = self.conv11(x)
        x = x.view(batch,-1)

        return x

```

```python

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
import torchvision
from torchvision import transforms

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import datetime
import time
import cv2
import subprocess
from sklearn.model_selection import train_test_split
from tqdm import tqdm
def getTemp():
    cmd = 'nvidia-smi', '--query-gpu=temperature.gpu', '--format=csv,noheader'
    temperature = int(subprocess.check_output(cmd))
    return temperature




class DataReader():
    def __init__(self,sensor,base_dir=DATA_DIR1,phase='train'):
        self.base_dir = base_dir
        print(base_dir)
        self.filename = os.listdir(self.base_dir)
        self.sensor = sensor
        self.phase = phase
    @property
    def data_size(self):
        return len(self.filename)
    def read(self,filename):
        df = pd.read_csv(os.path.join(self.base_dir,filename),usecols=self.sensor)
        df = df.astype(np.float32)
        # df = pd.concat([df[self.sensor],df[self.sen]],axis=1)
        return df
    def reserve(self):
        if self.phase == 'train':
          print(self.data_size)

          for i in range(self.data_size):
              X=np.array(self.read(self.filename[i]))[200:328,0]
              # print(X.shape)
              if i ==0:
                train_list = np.zeros((self.data_size,X.shape[0]))

              train_list[i] = X
              # x0.append(X)
          train_list,valid_list=train_test_split(train_list,test_size=0.1,shuffle=False,random_state=42)
          return train_list,valid_list
        else:
          print(self.data_size)
          print(self.base_dir)
          for i in range(self.data_size):
              X=np.array(self.read(self.filename[i]))[200:328,0]
              # print(X.shape)
              if i ==0:
                test_list = np.zeros((self.data_size,X.shape[0]))

              test_list[i] = X
              # x0.append(X)
          # train_list,valid_list=train_test_split(data_list,test_size=0.1,shuffle=False,random_state=42)
          return test_list
        # 全データをx0としてメモリーに読み込む

def add_noise(data):
    """ガウシアンフィルタを適用する。
    """
    return data + torch.normal(0,0.01, size=(data.shape))

class torch_dataset(Dataset):
    def __init__(self,X,y,path,phase='train'):
      if not os.path.exists('{}mean.npy'.format(path)):
        mean_ = np.mean(X)
        np.save('{}mean.npy'.format(path),mean_)
      else:
        mean_ = np.load('{}mean.npy'.format(path))
      if not os.path.exists('{}std.npy'.format(path)):
        std_ = np.std(X)
        np.save('{}std.npy'.format(path),std_)
      else:
        std_ = np.load('{}std.npy'.format(path))

      X = (X-mean_) / std_
      X = torch.from_numpy(X.astype(np.float32)).clone()

      self.transform = transforms.Lambda(add_noise)
      self.X = X
      self.phase = phase
      self.y = y
      self.data_num = len(self.X)
      # self.alpha = alpha

    def __len__(self):
        return self.data_num

    def __getitem__(self, idx):
        input_data = self.X[idx]
        if self.phase == 'train':
          input_data = self.transform(input_data)
          return input_data
        elif self.phase == 'test':
          label = self.y[idx]
          return input_data, label
        else:
          return input_data



output_path = './result/'
os.makedirs(output_path, exist_ok=True)
normal_data_reader = DataReader(SENSOR,DATA_DIR2,'test')
anomaly_data_reader = DataReader(SENSOR,DATA_DIR3,'test')
normal_test  = normal_data_reader.reserve()
anomaly_test  = anomaly_data_reader.reserve()
normal_test = torch_dataset(normal_test,np.zeros((len(normal_test))),output_path,'test')
anomaly_test = torch_dataset(anomaly_test,np.ones((len(anomaly_test))),output_path,'test')
normal_dataloader = DataLoader(normal_test,
                              batch_size=1,
                              shuffle=False,
                              drop_last=False)#データローダーへ
anomaly_dataloader = DataLoader(anomaly_test,
                              batch_size=1,
                              shuffle=False,
                              drop_last=False)

dataloaders_dict = {"normal": normal_test, "anomaly": anomaly_test}#辞書にまとめる
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print("使用デバイス：", device)
# model = IDCNN() #モデル作成
model = CAE()
# model = models.MobileNetV2(sample_rate=14000,window_size=1024,hop_size=320,mel_bins=64,fmin=50,fmax=8000,classes_num=8)
model.to(device)
model.load_state_dict(torch.load('{}/model.pth'.format(output_path)))
criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)
epochs = 1000
anomaly_score = {"normal":[], "anomaly":[]} # 学習曲線用
torch.backends.cudnn.benchmark = True
min_ = 1010000100

# print('Epoch：{}/{}'.format(epoch+1, epochs))

for phase in ['normal', 'anomaly']:
    model.eval()

    epoch_loss = 0.0
    discriminator_loss = 0.0

    for cnt,(inputs,label) in enumerate(tqdm(dataloaders_dict[phase])):
        inp    = inputs.clone()
        # print(inp.shape)
        # targ = target.clone()

        inp = inp.float().to(device)
        inp = inp.view(1,-1)



        out = model(inp)
        loss = criterion(out, inp)

        anomaly_score[phase].append(loss.item())

```
